Importing the needed libraries.

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import seaborn as sns
import visualizations
```

Loading the Dataset via pandas.

```python
df = pd.read_csv("Donnees_Lettres/20211116_Constance_de_Salm_Korrespondenz_Inventar_Briefe.csv", sep=';', encoding='utf8')
list(df)
```

```python
df['Nr.']
```

Extracting just the letters which were written by the same hand.

```python
search = [r'CdS_16.*', r'CdS_17.*', r'CdS_26.*', r'CdS_27.*', r'CdS_28.*', r'CdS_29.*', r'CdS_30.*', r'CdS_31.*']
df_part = pd.DataFrame()
for i, v in df.iterrows():
    found = False
    for s in search:
        key = v['Nr.'].replace('/', '_')
        if re.search(s, key) is not None:
            found = True
            break
    if found:
        df_part = df_part.append(v)
df_part
```

Exporting the extracted meta-Informations in "Donnes_Lettres"

```python
# !pip install openpyxl
df_part.to_csv('Donnees_Lettres/Constance_de_Salm_Special_lettres.csv', encoding='utf8', sep=';')
df_part.to_excel('Donnees_Lettres/Constance_de_Salm_Special_lettres.xlsx', encoding='utf8')

df_part
```

Visualizing the "Schlagwörter" of the special df.

```python
#!pip install wordcloud
import wordcloud as wc

def create_wordcloud(text):
    if type(text) is pd.Series:
        text: pd.Series
        tmp = ''
        for i, v in text.iteritems():
            try:
                for j in v.split('; '):
                    tmp += j + ' '
            except AttributeError:
                continue
        text = tmp
    cloud = wc.WordCloud(width=800, height=400, background_color="white")
    cl = cloud.generate_from_text(text)
    return cl.to_image()
create_wordcloud(df_part['Schlagwörter'])
```

Tracing specific keywords over time.

```python
def visualize_keywords(data: pd.DataFrame, keywords):
    vis = {}
    for i,v  in data.iterrows():
        try:
            date = v["Datierung (JJJJ-MM-TT)"].split('-')
        except AttributeError:
            # if v["Datierung"] is nan
            continue
        year = int(date[0])
        tmp = [year]
        vis[year] = {k: 0 for k in keywords} if year not in vis.keys() else vis[year]
        for k in keywords:
            k: str
            try:
                if re.search(k, v["Schlagwörter"]):
                    vis[year][k] += 1
            except TypeError:
                # if v["Schlagwörter"] is nan
                continue
    vis_df = pd.DataFrame(vis)
    sns.set_theme(style='whitegrid')
    plt.figure(figsize=[12, 6])
    for i, v in vis_df.iterrows():
        sns.scatterplot(x=[j for j, val in v.iteritems()], y=[val for j, val in v.iteritems()], label=i)
    plt.xlabel("Year")
    plt.ylabel("Usage")
    plt.legend()
    plt.show()

visualize_keywords(df_part, ['Prozess', 'Wirtschaft', 'Jugend'])
```
