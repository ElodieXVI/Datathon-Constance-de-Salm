# Constance de Salm #

Importing the needed libraries.

```python
import pandas as pd
import re
```

Loading the Dataset via pandas.

```python
df = pd.read_csv("Données archivage des lettres/20211116_Constance_de_Salm_Korrespondenz_Inventar_Briefe.csv", sep=';', encoding='utf8')
list(df)
```

```python
df['Nr.']
```

```python
search = [r'CdS_16.*', r'CdS_17.*', r'CdS_26.*', r'CdS_27.*', r'CdS_28.*', r'CdS_29.*', r'CdS_30.*', r'CdS_31.*']
df_part = pd.DataFrame();
for i, v in df.iterrows():
    found = False
    for s in search:
        key = v['Nr.'].replace('/', '_')
        if re.search(s, key) is not None:
            found = True
            break
    if found:
        df_part = df_part.append(v)
df_part
```
